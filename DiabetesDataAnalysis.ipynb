### Looking at Glucose Levels

```python
# Find the unique values of Glucose
df['Glucose'].unique()
```

```python
# List the frequency of values in Glucose
df['Glucose'].value_counts()
```

```python
# Select a random sample of 40 individuals from the population and assign the created numpy array to the variable sample.
sample = np.random.choice(df['Glucose'], 40)
```

> This will select 40 random people from Glucose

```python
# Calculate the sample mean and assign the value to the variable sample_mean.
sample_mean = np.mean(sample)
sample_mean 
```

> Calculated the mean glucose levels from those people

```python
# Calculate the standard deviation of the sample and assign the value to the variable sample_std.
sample_std = np.std(sample)
sample_std
```

>The values deviated from the mean by 33.656 sugar levels.

```python
# Calculate the standard error of the sample and assign it to the value sample_se.
sample_se = sample_std / np.sqrt(len(sample))
sample_se
```

> Calculated for standard error

```python
# Calculate for the confidence coeffcient for a 95% confidence level given the sample size and degrees of freedom, and assign the value to the variable c.
c = 2.021
c
```

```python
# Calculate the lower bound of the confidence interval for the sample mean and assign it to the variable lower_bound.
lower_bound = sample_mean - c * sample_se
lower_bound
```

```python
# Calculate the upper bound of the confidence interval for the sample mean and assign it to the variable upper_bound.
upper_bound = sample_mean + c * sample_se
upper_bound
```

```python
# Create a graph that compares Glucose against Outcome
sns.boxplot(x='Outcome', y='Glucose', data=df)
```

>This box and whisker plot shows that higher glucose levels are correlated to the patients having diabetes.
